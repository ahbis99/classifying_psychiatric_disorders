```python
import nibabel as nib
import numpy as np
import pandas as pd
import os
```

# Behavior Source Data

```python
# Reading
behavior_path = '../../data/BSNIP_behavior/behavior/'

behavior_files = os.listdir(behavior_path)

behavior_source = pd.read_csv(behavior_path+behavior_files[0], sep='\t')
for behavior_file in behavior_files[1:]:
    curr_behavior_source = pd.read_csv(behavior_path+behavior_file, sep='\t')
    behavior_source = pd.concat([behavior_source, curr_behavior_source], axis=0)
```

```python
behavior_source
```

```python
behavior_source.to_csv('../../data/processed/behavior_source.csv', index=False)
```

# Processing new GBC Cole-Antisevic parcellation data

```python
import os
data_path = 'BSNIP_cole_anticevic_2/pscalar/'
data_id = data_path.replace('/', '_')[:-1]
file_list = os.listdir(f'../../data/{data_path}')
current_data_session_id = [f.split('.')[0] for f in file_list]
```

```python
current_data_session_id
```

```python
# Concatenating Cole Anticevic data with previous data

import pickle
def get_file_data(file):
    return nib.load(f'../../data/{data_path}' + file).get_fdata()[0]

if os.path.exists(f'../../data/pickles/{data_id}.pickle'):
    with open(f'../../data/pickles/{data_id}.pickle', 'rb') as f:
        pscalar_data_array = pickle.load(f)
else:
    pscalar_data_array = np.array([get_file_data(f) for f in file_list])
    with open(f'../../data/pickles/{data_id}.pickle', 'wb') as f:
        pickle.dump(pscalar_data_array, f)
```

```python
pscalar_data_array.shape
```

```python
# Naming the columns
column_names = [f"X{i}" for i in range(1, 13)]
pscalar_data = pd.DataFrame(pscalar_data_array, columns=column_names)

# Adding file identifiers
pscalar_data['session_id'] = current_data_session_id
pscalar_data.head()
```

```python
# Merging new Cole Anticevic data with behavior data to get the groups
data = pd.merge(behavior_source[['session_id', 'Group']], pscalar_data, on='session_id', how='inner')
data.to_csv(f'../../data/processed/pscalar/{data_id}.csv', index=False)
data
```

```python
neural_gbc_source = pd.read_csv('../../data/BSNIP_data/BSNIP_data/neural/neural_parcellated_gbc.csv')
behavior_original_source = pd.read_csv('../../data/BSNIP_data/BSNIP_data/behavior/behavior_data.csv')

# Selecting only PANSS columns and renaming the first column to 'id'
behavior_original_source.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
PANSS_selector = behavior_original_source.columns[behavior_original_source.columns.str.contains('PANSS')]
PANSS_selector = np.append(['id'], PANSS_selector)
print(PANSS_selector)
behavior_original_source = behavior_original_source[PANSS_selector]

# Merging the two dataframes
neural_gbc_source.rename(columns={'Unnamed: 0': 'id'}, inplace=True)
neural_gbc_nn = neural_gbc_source.merge(behavior_original_source, on='id', how='inner')
display(neural_gbc_nn)
neural_gbc_nn.to_csv('../../data/processed/pscalar/neural_gbc_nn.csv', index=False)
```
